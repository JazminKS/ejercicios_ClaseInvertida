5. Si nos fijamos, en la primera fila antes solo teníamos un paréntesis, pero los hemos eliminado en el paso anterior, y ahora tenemos un string vacío, lo cual no tiene mucho sentido. En este ejercicio debereis reemplazar esos strings vacios por nulos de Pandas (NOTA estos deben ser indicados de la siguiente forma np.nan, y deberéis importar numpy).

```python
mergeado_izq.head(2)
```

```python
def cambiar_vacio(cadena):
    """Reemplaza cadenas vacías con np.nan, dejando otros valores sin cambios."""
    if isinstance(cadena, str) and cadena == "":
        return np.nan
    else:
        return cadena 
```

```python
mergeado_izq["merchant_info_subtitle"] = mergeado_izq["merchant_info_subtitle"].apply(cambiar_vacio)
```

```python
mergeado_izq[["merchant_info_subtitle", "merchant_rating_count"]] 
```
