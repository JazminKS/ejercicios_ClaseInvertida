```python
# importamos las libreras que necesitamos

# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd
import numpy as np

# Imputacin de nulos usando mtodos avanzados estadsticos
# -----------------------------------------------------------------------
from sklearn.impute import SimpleImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.impute import KNNImputer

# Libreras de visualizacin
# -----------------------------------------------------------------------
import seaborn as sns
import matplotlib.pyplot as plt
# Configuracin
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

# Ejercicios

1. Cargad el csv.

```python
df = pd.read_csv("merchant_ejercicios_apply.csv", index_col = 0)
df.head(2)
```

2. Qu columnas tienen nulos?

```python
colum_nulos = df.isnull().sum() 
colum_nulos
```

```python
df_nulos = pd.DataFrame(colum_nulos, columns = ["cantidad_nulos"])

df_nulos[df_nulos["cantidad_nulos"] > 0]
```

3. Explorad los nulos que hay en el DataFrame.

```python
# Obtenemos la lista de columnas categricas que tienen nulos
nulos_esta_cat = df[df.columns[df.isnull().any()]].select_dtypes(include = "O").columns
print("Las columnas categricas que tienen nulos son : \n ")
print(nulos_esta_cat)
```

```python
for col in nulos_esta_cat:
    print(f"La distribucin de las categoras para la columna {col.upper()}")
    display(df[col].value_counts() / df.shape[0])  # display es una funcin utilizada para mostrar objetos de manera ms legible en Jupyter Notebooks o entornos similares. 
    print("........................")
```

```python
columnas_moda = ["product_color", "product_variation_size_id", "origin_country"]
columnas_desconocido = ["merchant_name", "merchant_info_subtitle"]
```

```python
for columna in columnas_moda:
    
    # calculamos la moda para la columna por la que estamos iterando
    moda = df[columna].mode()[0]    
    
    # utilizando el mtodo fillna reemplazamos los valores nulos por la moda calculada en el paso anterior. 
    df[columna] = df[columna].fillna(moda)

# por ltimo chequeamos si se han eliminado los nulos en las columnas de "marital" y "loan"
print("Despus del reemplazo usando 'fillna' quedan los siguientes nulos")

df[columnas_moda].isnull().sum()
```

```python
for columna in columnas_desconocido:
    
    # reemplazamos los nulos por el valor Unknown para cada una de las columnas de la lista
    df[columna] = df[columna].fillna("Unknown")
    
# comprobamos si quedan nulos en las columnas categricas. 
print("Despus del reemplazo usando 'fillna' quedan los siguientes nulos")
df[columnas_desconocido].isnull().sum()
```

```python
nulos_esta_num = df[df.columns[df.isnull().any()]].select_dtypes(include = np.number).columns
print("Las columnas numricas que tienen nulos son : \n ")
print(nulos_esta_num)
```

```python
for col in nulos_esta_num:
    print(f"La distribucin de las categoras para la columna {col.upper()}")
    display(df[col].value_counts() / df.shape[0])  # display es una funcin utilizada para mostrar objetos de manera ms legible en Jupyter Notebooks o entornos similares. 
    print("........................")
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 3, figsize = (20,10)) 

axes = axes.flat

for indice, col in enumerate(nulos_esta_num):
    sns.boxplot(x = col, data = df, ax = axes[indice])
    
plt.tight_layout()
fig.delaxes(axes[-1]);
```

```python
df_copia = df.copy()
df_copia.head(1)
```

```python
imputer_iterative = IterativeImputer(max_iter = 20, random_state = 42) #max_iter=20 especifica el nmero mximo de iteraciones/ random_state=42 asegura la reproducibilidad de los resultados.

# fit_transform() ajusta el modelo IterativeImputer a los datos seleccionados y luego transforma los datos imputando los valores faltantes.
imputer_iterative_imputado = imputer_iterative.fit_transform(df_copia[['rating_five_count', 'rating_four_count', 'rating_three_count','rating_two_count', 'rating_one_count']])

# comprobamos que es lo que nos devuelve, que en este caso es un array tambin
imputer_iterative_imputado
```

```python
df_copia[['rating_five_count', 'rating_four_count', 'rating_three_count','rating_two_count', 'rating_one_count']] = imputer_iterative_imputado

# comprobamos los nulos
print(f"Despus del 'Iterative' tenemos: \n{df_copia[['rating_five_count', 'rating_four_count', 'rating_three_count','rating_two_count', 'rating_one_count']].isnull().sum()} nulos")
```

```python
# hagamos ahora lo mismo con el KNNImputer. NOTA, la ejecucin de esta celda puede tardar un ratito

# instanciamos la clase del KNNImputer
imputer_knn = KNNImputer(n_neighbors = 5)

# ajustamos y transformamos los datos
imputer_knn_imputado = imputer_knn.fit_transform(df_copia[['rating_five_count', 'rating_four_count', 'rating_three_count','rating_two_count', 'rating_one_count']])

# comprobamos que es lo que nos devuelve, que sigue siendo un array
imputer_knn_imputado
```

```python
# por ltimo nos queda aadir ese array al DataFrame como hemos hecho hasta ahora
df_copia[['rating_five_count', 'rating_four_count', 'rating_three_count','rating_two_count', 'rating_one_count']] = imputer_knn_imputado


# comprobamos los nulos
print(f"Despus del 'KNN' tenemos: \n{df_copia[['rating_five_count', 'rating_four_count', 'rating_three_count','rating_two_count', 'rating_one_count']].isnull().sum()} nulos")
```
