Contesta a las siguientes preguntas utilizando la visualización más apropiada.

1. ¿Cómo se distribuyen los precios de los productos?

```python
df["price"].value_counts()
```

```python
# creamos la gráfica de seaborn
sns.histplot(x = "price", 
             data = df, 
             color = "violet", 
             kde = True, 
             bins = 50 )

# usando 'plt.xlabel()' cambiamos el nombre del eje x
plt.xlabel("Precio")

# usando el método 'plt.ylabel()' cambiamos el nombre del eje y
plt.ylabel("Conteo");
```

2. ¿Cuál es la relación entre el precio y la calificación promedio de los productos?

```python
sns.regplot(x = "price", 
            y = "mean_product_ratings_count", 
            data = df, 
            marker = "d", 
            line_kws = {"color": "black", "linewidth": 1}, # cambiamos el color y el grosor de la linea de tendencia
            scatter_kws = {"color": "blue", "s": 1} # cambiamos el color y el tamaño de los puntos del scaterplot
            )

# cambiamos los nombres de los ejes como hemos estado haciendo hasta ahora
plt.xlabel("Precio")
plt.ylabel("Calificación promedio de los productos")

# ponemos título a la gráfica
plt.title("Relación calificación promedio de los productos y precio")

# quitamos la linea de arriba y de la derecha
plt.gca().spines['right'].set_visible(False) # quitamos la línea de la derecha
plt.gca().spines["top"].set_visible(False) # quitamos la línea de arriba;
```

3. ¿Cómo se distribuyen los precios de los productos en función de si se utilizan anuncios promocionales (uses_ad_boosts) o no?

```python
# Gráfico de barras
sns.barplot(x = "uses_ad_boosts", 
            y = "price",
            data=df)

plt.xlabel("Anuncios promocionales")
plt.ylabel("Precio")

plt.title("Relación entre anuncios promocionales y precio")

plt.show()
```

4. ¿Existe una relación entre el uso de anuncios y las unidades vendidas?

```python
sns.stripplot(x = "uses_ad_boosts", 
              y = "units_sold",
              data=df,
              size= 3,
              marker = "v",
              color = "orange")

plt.xlabel("Anuncios promocionales")
plt.ylabel("Unidades vendidas")

plt.title("Relación entre anuncios promocionales y unidades vendidas")

plt.show()
```

5. ¿Cuál es la relación entre el precio de envío y las unidades vendidas?

```python
sns.regplot(x = "shipping_option_price", 
            y = "units_sold", 
            data = df, 
            marker = "d", 
            line_kws = {"color": "black", "linewidth": 1}, # cambiamos el color y el grosor de la linea de tendencia
            scatter_kws = {"color": "orange", "s": 1} # cambiamos el color y el tamaño de los puntos del scaterplot
            )

# cambiamos los nombres de los ejes como hemos estado haciendo hasta ahora
plt.xlabel("Precio de envío")
plt.ylabel("Unidades vendidas")

# ponemos título a la gráfica
plt.title("Relación entre el precio de envío y las unidades vendidas")

# quitamos la linea de arriba y de la derecha
plt.gca().spines['right'].set_visible(False) # quitamos la línea de la derecha
plt.gca().spines["top"].set_visible(False) # quitamos la línea de arriba;
```

7. Compara diferentes distribuciones de precios de productos en función de si tienen el distintivo "Producto local" o "Buena calidad del producto". Tendréis que crear dos gráficas en un subplot, en la primera poner los precios de los productos locales ("badge_local_product")

```python

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 5))

# Primer subplot
sns.barplot(x="badge_product_quality",
            y="price",
            data=df,
            color="orange",
            ax=axes[0])

axes[0].set_xlabel("Distintivo calidad del producto")
axes[0].set_ylabel("Precio")
axes[0].set_title("Relación entre distintivo calidad del producto y precio")
axes[0].spines['right'].set_visible(False)
axes[0].spines['top'].set_visible(False)

# Segundo subplot
sns.barplot(x="badge_local_product",
            y="price",
            data=df,
            color="pink",
            ax=axes[1])

axes[1].set_xlabel("Distintivo de producto local")
axes[1].set_ylabel("Precio")
axes[1].set_title("Relación entre distintivo de producto local y precio")
axes[1].spines['right'].set_visible(False)
axes[1].spines['top'].set_visible(False)

plt.show()  # Agrega esta línea para mostrar el gráfico
```
